```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/multiple_linear_regression_dataset.csv')
df
```

```python
df.info()
df.columns
```

```python
x = df[['age', 'experience']]
y = df['income']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
#y_train = np.array(y_train).reshape(-1,1)
#y_test = np.array(y_test).reshape(-1,1)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

plt.plot(x_test, y_test, color='black',label='Actual value')
plt.plot(x_test, y_pred, color='blue', linewidth=2,label='Predicted value')
plt.xlabel('Experience')
plt.ylabel('Income')
plt.title('Experience vs Income')
plt.legend()
plt.show()
```
